```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.orm import Session
import pandas as pd
import datetime
```

```python
# defining sqlite db and defining connection and session handlers
engine = create_engine("sqlite:///hawaii.sqlite" , echo=True)
conn = engine.connect()
sess = Session(bind=engine)
```

```python
# Import and establish Base for which classes will be constructed 
# Import modules to declare columns and column data types
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
from sqlalchemy import Column, Integer, String, Numeric, Text, Float, Date, DateTime ,UniqueConstraint, MetaData
import sqlalchemy
```

```python
#define data type and classes for tables

# ----------------------------------
class c_station(Base):
    __tablename__ = 'station'
    id = Column(Integer, primary_key=True, autoincrement=True)
    station = Column(String(15))
    name = Column(String(50))
    latitude = Column(Numeric(10,4))
    longitude = Column(Numeric(10,4))
    elevation = Column(Numeric(10,4))
    __table_args__ = (UniqueConstraint('station', name='uk_station'), )

# sqllite has datetime type by sqlalchemy does not support it with sqllite
class c_measurement(Base):
    __tablename__ = 'measurement'
    id = Column(Integer, primary_key=True, autoincrement=True)
    station = Column(String(15))
    date = Column(String(10))
    prcp = Column(Numeric(10,4))
    tobs = Column(Numeric(10,4))
    __table_args__ = (UniqueConstraint('station','date', name='uk_station_date'),)
# Create both the Surfer and Board tables within the database
Base.metadata.create_all(conn)
# --
```

```python
# loading clean csv data in dataframe
df_station = pd.read_csv('clean_hawaii_stations.csv')
df_measurement = pd.read_csv('clean_hawaii_measurements.csv')
df_measurement.sort_values(['date'], ascending=False).head(20)
```

```python
#---medata for table
metadata = MetaData(bind=engine)
metadata.reflect()

#---convert to json df_station
v_json = df_station.to_dict(orient='records')
v_t1 = sqlalchemy.Table('station', metadata, autoload=True)
#----now bulk insert df_station
conn.execute(v_t1.insert(), v_json)

#---convert to json df_station
v_json = df_measurement.to_dict(orient='records')
v_t1 = sqlalchemy.Table('measurement', metadata, autoload=True)
#----now bulk insert df_station
conn.execute(v_t1.insert(), v_json)

```
